# Acerca de los bucles while

Un ciclo while realiza una operaci√≥n mientras (while, en ingl√©s) una determinada condici√≥n es True(verdadera). Funcionan para evaluar si hay otra l√≠nea en un archivo, si se ha establecido una marca, si un usuario ha terminado de escribir valores o bien ha cambiado algo m√°s para indicar que el c√≥digo puede dejar de realizar la operaci√≥n.

Lo m√°s importante que se debe recordar al crear ciclos while es asegurarse de que cambia la condici√≥n. Si la condici√≥n siempre es True, Python seguir√° ejecutando el c√≥digo hasta que el programa se bloquee.

La sintaxis de un ciclo while es similar a la de una instrucci√≥n if. Proporciona una condici√≥n, as√≠ como el c√≥digo que quieres ejecutar mientras la condici√≥n sea True (Os√©a, mientras la condici√≥n se cumpla o sea verdadera).
En palabras simples, eres Patrick y te invitaron a un a fiesta de cumplea√±os, mientras tengas una invitacion podras entrar, si no la tienes no podras entrar, o bien no puede ser solo una invitacion, mientras tengas una invitacion podras entrar, si no solo tienes un regalo pero no tienes invitacion no puedes entrar

Un ciclo while tiene tres partes importantes:

* La palabra while, seguida de un espacio.

* La condici√≥n que se va a probar. Si la condici√≥n es True, se ejecutar√° el c√≥digo dentro del ciclo while.

* El c√≥digo que quiere ejecutar para cada elemento del objeto iterable, seguido de espacios en blanco anidados

```python
x = 9
while x > 0:
    x -= 1
    print(x)
```

```python
y = ["Uno", "Dos", "Tres"]
while y:
    y.pop(0)
    print(y)  
```

```python
# Creamos la variable que almacena el texto
user_input = ''
# Creamos la lista que almacena cada uno de los textos que el usuario ingresa, una lista vacia
inputs = []

# La funci√≥n LOWER devuelve una cadena de caracteres que presenta el mismo tipo de datos que la cadena de entrada
while user_input.lower() != 'listo':     
    # Verificamos si hay un valor en user_input
    if user_input:
        # Almacenamos ese valor en la lista
        inputs.append(user_input)
    # Capturamos un nuevo valor
    user_input = input('Ingresa un nuevo valor, textea lsito cuando hayas terminado, el valor asigando: ')

inputs    
```

Observa la instrucci√≥n if dentro del ciclo while. Esta instrucci√≥n prueba si hay un valor de cadena dentro de user_input. Si el ciclo while se ejecuta por primera vez, no hay ning√∫n valor, por lo que no hay nada que almacenar en inputs

# Sentencia for con listas 
Los ciclos for (o ciclos para) son una estructura de control c√≠clica. Nos permiten ejecutar una o varias l√≠neas de c√≥digo de forma iterativa (o repetitiva), pero teniendo cierto control y conocimiento sobre las iteraciones. En el ciclo for en Python, es necesario tener un valor inicial y un valor final, y opcionalmente podemos hacer uso del tama√±o del "paso" entre cada "giro" o iteraci√≥n del ciclo

```python
Cuenta = [5, 4, 3, 2, 1]

for n in (Cuenta):
  print(n)
print('Explote, termino la cuenta regresiva')  
```

La sentencia **for** es una instrucci√≥n con partes importantes:

* La palabra for, seguida de un espacio.
* El nombre de la variable que quieres crear para cada valor de la secuencia (n en este ejmplo) se pueden declarar variables.
* La palabra in, entre espacios.
* El nombre de la lista (Cuenta en el ejemplo utilizado) u objeto iterable que quieres recorrer en ciclo, seguido de dos puntos (:) la identacion.
* El c√≥digo que quieres ejecutar para cada elemento del objeto iterable, separado por espacios en blanco anidados.
Vamos a cambiar el c√≥digo anterior para que espere un segundo entre cada n√∫mero mediante la funci√≥n sleep():

```python
# De la biblioteca time, importamos (traemos) la clase sleep
from time import sleep

# Creamos una lista [VALORES]
Cuenta = [5, 4, 3, 2, 1]

# Para cada n√∫mero(depende de que tipo sea, en eset caso es int) en Cuenta
for n in Cuenta:
    #Muestra el n√∫mero
    print(n)
    # Espera (2 segundos)
    sleep(2)

# Muestra el mensaje, tener en cuenta la identacion 
print("Kabooom!")
```

```python
# Iniciamos el contador en cero
Contador = 0 


for i in range(1000):
    if (i % 33 == 0): # Preguntamos si el residuo es 0 (es m√∫ltiplo de 33)
        Contador += 1 # Si es m√∫ltiplo aumentamos el contador en 1
    
    # Si no es m√∫ltiplo no hacemos nada

#Mostramos el valor del contador
print(Contador)

#Este ciclo for nos permitir√° saber que existen 31 m√∫ltiplos del n√∫mero 33 en los n√∫meros del 0 al 1000 (incluido el 1000 mismo)
```

# Palabra Break en los ciclos
Python break es una sentencia que permite parar un bucle por completo en cuanto se da o deja de darse una condici√≥n externa

```python
for letra in 'Marinita':
    if letra == 'i':
        print(f'Letra encontrada: {letra}')
        
else:
    print('Fin ciclo for') #Observese la identacion, el else esta con el cilco for, no con el if 
```

```python
for letra in 'Marinita':
    if letra == 'i':
        print(f'Letra encontrada: {letra}')
        break # como enontro la primera i rompera nuestro ciclo for, inlusive rompe el else
        
else:
    print('Fin ciclo for') #Observese la identacion, el else esta con el ciclo for, no con el if 

    #esto es util para encontrar un elemneto de una lista de datos y asi no correr un for infnito 
```
