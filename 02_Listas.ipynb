# Agregar valores a una lista 
Las listas en python son dinamicas puede agregar o quitar elemtnos de las mismas, para agregar un elemento a una lista usaremos la sentencia .append(value)

Por ejemplo agregar el siguiente planeta a nuestra listas de planetas, agregarlo al final: Krypton

```python
Mi_lista_de_planetas.append('Krypton')
Mi_lista_de_planetas

numero_de_planetas = len(Mi_lista_de_planetas)
print('Actualmente el numero de planetas es :', numero_de_planetas, ',pero uno no es de nuestro sistema solar')
```

# Eliminar valores de una lista 

Puedes eliminar el √∫ltimo elemento de una lista llamando al m√©todo .pop() de la variable de lista

```python
Mi_lista_de_planetas.pop()  # Goodbye, Krypton
numero_de_planetas = len(Mi_lista_de_planetas)
print('Los planetas que hay en realidad son', numero_de_planetas, 'en nuestro sistema solar.')
numero_de_planetas
```

# Ordenar Listas
El m√©todo sort() nos permite ordenar una lista en orden ascendente o descendente

```python
numeros = [16, 4, 4, 9, 1, 3, 20, 8, 1, 2, 3, 4, 5, 0.5, 0.7, -1]
numeros.sort()
numeros
```

# Ordenar mayor a menor 

Par√°metro reverse para ordenar de mayor a menor.

Tanto al m√©todo sort() como a la funci√≥n sorted() se le puede indicar el par√°metro reverse para modificar el orden ascendente en que se ordena una lista. As√≠, si se indica reverse = bool

```python
numeros = [16, 4, 9, 1, 3, 20, 8]
numeros.sort(reverse = True)
numeros
```

```python
#En efrcto, sorted devuelve una nueva lista
numeros = [16, 4, 9, 1, 3, 20, 8]
ordenados = sorted(numeros, reverse = True)
ordenados
```

# if __name__ == '__main__'

Sabremos si el srcip es ejecutado desde la operacion principal, si eres de los que busca codigo encontaras mucho esa forma y si, al principio no se entiende para que se usa 



*   ¬øQu√© es y c√≥mo funciona?
*   ¬øCual es la diferencia?

En palabras mu simples 
Nos permite reusar c√≥digo como si fuera un m√≥dulo.
Mejor con un ejemplo



```python
# Es comun encontrarnos codigo de la forma: 
# Supongamos que este codigo lo guardamos en algo llamdo Potencia_al_cuadrado.py
def potencia_alcuadrado(var):
    return var * var

if __name__ == '__main__':
    print("20 al cuadrado es igual a:", potencia_alcuadrado(20))
```



1.   Cuando el int√©rprete lee un archivo de c√≥digo, ejecuta todo el c√≥digo global que se encuentra en √©l. Esto implica crear objetos para toda funci√≥n o clase definida y variables globales

2.   Todo m√≥dulo (archivo de c√≥digo) en Python tiene un atributo especial llamado __name__ que define el espacio de nombres en el que se est√° ejecutando. Es usado para identificar de forma √∫nica un m√≥dulo en el sistema de importaciones.

3.   m√≥dulo, el atributo __name__ pasa a contener el nombre del archivo en si.


4.  El int√©rprete pasa el valor del atributo __name__ a la cadena '__main__' si el m√≥dulo se est√° ejecutando como programa principal (cuando lo ejecutas llamando al int√©rprete en la terminal con python my_modulo.py, haciendo doble click en √©l, ejecut√°ndolo en el int√©rprete interactivo, etc ).


Es decir, si tienes un archivo llamado mi_modulo.py, si lo ejecutamos como programa principal el atributo __name__ ser√° '__main__', si lo usamos import√°ndolo desde otro m√≥dulo (import mi_modulo) el atributo __name__ ser√° igual a 'mi_modulo'.

B√°sicamente, lo que haces usando if __name__ == ‚Äú__main__‚Äù: es ver si el m√≥dulo ha sido ejecutado directamente o no (importado). Si se ha ejecutado como programa principal se ejecuta el c√≥digo dentro del condicional.

Una de las razones para hacerlo es que, a veces, se escribe un m√≥dulo (un archivo .py) que se puede ejecutar directamente pero que, alternativamente, tambi√©n se puede importar y reutilizar sus funciones, clases, m√©todos, etc en otro m√≥dulo
