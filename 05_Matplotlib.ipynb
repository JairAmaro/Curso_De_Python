# Conociendo Pyplot - - - fig , ax = plt.subplots()

El comando plt.subplots() crea 2 objetos: una Figure ***fig*** (figura) que es un contenedor de todo lo que ver√°s en el gr√°fico y los Axes ***ax*** (ejes) que contienen la informaci√≥n a graficar. 

Para poder visualizar los datos que tengamos, estos se deben a√±adir a los ejes. La manera de hacerlo es mediante m√©todos de ploteo que trae el objeto ***Axes***. El m√°s b√°sico de todos es ***plot***, el cual crea un gr√°fico lineal y requiere los datos del eje X y Y como argumentos


* **label = 'name'** ‚Üí pone una etiqueta al plot para saber a qu√© corresponde cada color (muy √∫til cuando vas a poner varias gr√°ficas en un solo plot).
* **marker = 'o'** ‚Üí a√±ade marcadores al plot y tambi√©n puedes indicar el tipo de marcador que quieres. [todas las figuras posibles](https://matplotlib.org/stable/api/markers_api.html)
* **linestyle = '--'** ‚Üí cambia el tipo de l√≠nea. As√≠ mismo, aqu√≠ podr√°s encontrar los distintos tipos de l√≠nea. Incluso si colocas **'None'**, simplemente ya no habr√° l√≠neas y quedar√°n solo los marcadores.
* **color = 'b'** ‚Üí cambia el color de tus datos. Aqu√≠ encontrar√°s c√≥mo poner cualquier color e incluso dejarlos transparentes con alpha.

Adicionalmente, el objeto ***Axes*** trae una serie de m√©todos que podr√≠as usar para que sea sencillo identificar de qu√© trata tu gr√°fica. Col√≥calos siempre antes de mostrar tu gr√°fica antes de plt.show().


* ***.set_xlabel('Name')*** ‚Üí coloca nombre al eje de las X.
* ***.set_ylabel('Name')*** ‚Üí coloca nombre al eje de las Y.
* ***.set_title('Name')*** ‚Üí coloca el t√≠tulo de tu plot.
* ***.set(xlabel='Name', ylabel='Name')*** ‚Üí tambi√©n puedes usar √∫nicamente set y pasar como par√°metros lo anterior.
* ***.legend()*** ‚Üí si anteriormente pusiste etiquetas con label, legend las mostrar√°.

# Step Demo
Este ejemplo demuestra el uso de pyplot.step para curvas constantes por tramos. En particular, ilustra el efecto del par√°metro.
* plot(x, y)         plot x , y utilizando el estilo y el color de l√≠nea predeterminados
* plot(x, y, 'bo')   plot x , y usando marcadores circulares azules
* plot(y)            plot y usando x como index de √≠ndice $0..N-1$
* plot(y, 'r+')      ditto, pero con puntos rojos

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(14) #np.array(lista) : Crea un array a partir de la lista o tupla lista y devuelve una referencia a √©l
y = np.sin(x / 2) # el seno nuestro arreglo/2

# Curva verde
plt.step(x, y + 2, label='')
plt.plot(x, y + 2, 'o--', color='grey', alpha=0.5)

# Curva naranja
plt.step(x, y + 1, where = 'mid', label='mid') # where 'pre', 'post', 'mid'
plt.plot(x, y + 1, 'o--', color='purple', alpha=0.3)

# Curva azul
plt.step(x, y, where = 'post', label='post')
plt.plot(x, y, 'o--', color='brown')

# Ploteamos
plt.grid(axis='x', color='0.95')
plt.legend(title = 'Parametros:')
plt.title('Parametros')
plt.show()
````

# Interpolaci√≥n
La funci√≥n np.interp de NumPy permite obtener una base comun de tiempos para realizar operaciones matem√°ticas entre datos adquiridos


````python
# La interpolaci√≥n sirve para poder hacer operaciones entre tiras de datos
# distintas que no comparten escalas de tiempos (entro otras cosas)


from numpy import *
import matplotlib.pyplot as plt


# Datos originales que no comparten base de tiempos
tiempo1 = linspace(0,4*pi, 50 ) # un arreglo de 50 datos, incia en 0 termina en 4*pi
datos1  = cos(tiempo1)


tiempo2 = linspace(0,4*pi, 20 ) # los mismo pero 33 datos
datos2  = sin(tiempo2)


plt.figure()

# Graficamos los datos
plt.subplot(2,1,1)

plt.plot( tiempo1 , datos1 , '.-' )
plt.plot( tiempo2 , datos2 , '.-' )

# Se√±alamos las posiciones temporales de datos2
plt.gca().set_xticks( tiempo2 ,minor=True)
plt.grid(b=True ,linestyle = ':', color = 'r',which = 'major' ,  alpha= 0.5) # major max y minor menor



plt.subplot(2,1,2)

# Creamos nuevos vectores de datos que interpolando en forma lineal en las posiciones
# en que los valores no est√°n definidos

tiempo_general = linspace(0,4*pi, 100 )
datos1_general = interp( tiempo_general , tiempo1, datos1 )
datos2_general = interp( tiempo_general , tiempo2, datos2 )

# Ahora que los vecotres tiene la misma base de tiempos, podemos hacer
# operaciones entre ellos
plt.plot( tiempo_general , datos1_general                , '-' , label='datos1' )
plt.plot( tiempo_general , datos2_general                , '-' , label='datos2'  )
plt.plot( tiempo_general , datos1_general-datos2_general , '-' , label='datos1-datos2', lw=3  )
plt.legend(title = 'Parametros:')


plt.gca().set_xticks( tiempo2 ,minor=True)
plt.grid(b=True,linestyle=':', color='lightgray')


````

En matem√°ticas, un sistema de ecuaciones param√©tricas permite representar una curva o superficie en el plano o en el espacio, mediante valores que recorren un intervalo de n√∫meros reales, mediante una variable, llamada par√°metro, considerando cada coordenada de un punto como una funci√≥n dependiente del par√°metro.

La representaci√≥n param√©trica de una curva en un espacio n-dimensional consiste en n funciones de una variable t que en este caso es la variable independiente o par√°metro (habitualmente se considera que t es un n√∫mero real y que los puntos del espacio n-dimensional est√°n representados por n coordenadas reales), de la forma

 $$x_i = f_i (t) : [a,b] ‚Üí\mathbb{R}$$ 
 donde $x_i$ representa la $i$-√©sima  coordenada del punto generado al asignar valores del intervalo $[a, b]$ a $t$. Por ejemplo, para representar una curva en el espacio se usan 3 funciones $x = x(t), y = y(t), z = z(t)$

 Se dice que un punto de la curva correspondiente a un valor t del intervalo es un punto ordinario si las derivadas de las funciones param√©tricas existen en y son continuas en ese punto y al menos una es distinta de 0. Si un arco de curva est√° compuesto solamente de puntos ordinarios se denomina suave.

 Es com√∫n resumir las ecuaciones param√©tricas de una curva en una sola ecuaci√≥n vectorial

 $$\vec{r}(t) = \sum_{k = 1}^{n} f_k (t) e_k = f_1 (t)e_1 + f_2 (t)e_2 + \cdots + f_n (t) e_n \in \mathbb{R}^n $$

 donde $e_k$ representa al vector unitario correspondiente a la coordenada ${\displaystyle k}$-√©sima. Por ejemplo, las funciones param√©tricas de un c√≠rculo unitario con centro en el origen son $x = cos t, y = sen t$. Podemos reunir estas ecuaciones como una sola ecuaci√≥n de la forma

 $$\vec{r}(t) = cos(t)i + sen(t)j$$

 siendo ${\displaystyle \scriptstyle \left\{{\hat {i}},{\hat {j}}\right\}}$ la base usual del espacio bidimensional real.

````python
fig = plt.figure()
ax = fig.gca(projection='3d')


theta = np.linspace(-3 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label = 'Curva parametrica' )
ax.legend()


plt.show()
````

# ¬øQu√© es una botella de Klein?

Es un ejemplo de ***superficie*** (variedad topol√≥gica de dimensi√≥n 2) ***compacta*** (cerrada y acotada), ***conexa*** (de una pieza) y no orientable (contiene bandas de M√∂bius).

vamos a dejarlo en que ser√≠a una ¬´botella¬ª que no tiene ni interior ni exterior, y tiene una sola cara. Digamos que se podr√≠a recorrer en su totalidad de forma continua

Sea la siguiente reggla de correspodencia 

$ x= \left\{ \begin{array}{lcc}
             6 \cdot cos(u)\cdot(1 + sen(u)) + r \cdot cos(v + \pi)   \\
             \\ (((6 \cdot cos (u) \cdot (1 + sen(u)) + r \cdot cos(u) \cdot cos(v)) &  si & 0 \leq u < \pi 
             \end{array}
   \right. $

$y = \left\{ \begin{array}{lcc}
             16 \cdot sen(u)   \\
             \\ (16 \cdot sen(u) + r \cdot sen(u) \cdot cos(v)  ) &  si & 0 \leq u < \pi 
             \end{array}
   \right.$

Sea 
$$ r = 4 \cdot (1 - \frac{cos(u)}{2})  $$

$$ z = r \cdot sen(v) $$

Veamos que pasa si graficamos 



````python
import mpl_toolkits.mplot3d.axes3d as axes3d
import matplotlib.pyplot as plt
import numpy as np

coseno = np.cos
seno = np.sin
pi = np.pi

def botella(u, v):
    mitad = (0 <= u) & (u < pi)
    r = 4*(1 - coseno(u)/2)
    x = 6*coseno(u)*(1 + seno(u)) + r*coseno(v + pi)
    x[mitad] = ((6*coseno(u)*(1 + seno(u)) + r*coseno(u)*coseno(v))[mitad])
    y = 16 * seno(u)
    y[mitad] = (16*seno(u) + r*seno(u)*coseno(v))[mitad]
    z = r * seno(v)
    return x, y, z

u, v = np.linspace(0, 2*pi, 30), np.linspace(0, 2*pi, 30)      #np.linspace (   , tama√±o de las lineas de la botellas por pi, )
ux, vx =  np.meshgrid(u,v)  #Rejas
x, y, z = botella(ux, vx)  # usamos la funcion

#Ploteamos
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
plot = ax.plot_surface(x, y, z,
                        rstride = 1, 
                        cstride = 1,
                        cmap = plt.get_cmap('inferno'),
                        linewidth = 0,
                        antialiased = True)
ax.set_title('Botella de Klein');

plt.show()
````

````python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


def plot_colortable(colors, title, sort_colors=True, emptycols=0):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12
    topmargin = 40

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(color))),
                         name)
                        for name, color in colors.items())
        names = [name for hsv, name in by_hsv]
    else:
        names = list(colors)

    n = len(names)
    ncols = 4 - emptycols
    nrows = n // ncols + int(n % ncols > 0)

    width = cell_width * 4 + 2 * margin
    height = cell_height * nrows + margin + topmargin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-topmargin)/height)
    ax.set_xlim(0, cell_width * 4)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()
    ax.set_title(title, fontsize=24, loc="left", pad=10)

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        swatch_end_x = cell_width * col + swatch_width
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.hlines(y, swatch_start_x, swatch_end_x,
                  color=colors[name], linewidth=18)

    return fig

plot_colortable(mcolors.BASE_COLORS, "Base Colors",
                sort_colors=False, emptycols=1)
plot_colortable(mcolors.TABLEAU_COLORS, "Tableau Palette",
                sort_colors=False, emptycols=2)

#sphinx_gallery_thumbnail_number = 3
plot_colortable(mcolors.CSS4_COLORS, "CSS Colors")

# Optionally plot the XKCD colors (Caution: will produce large figure)
#xkcd_fig = plot_colortable(mcolors.XKCD_COLORS, "XKCD Colors")
#xkcd_fig.savefig("XKCD_Colors.png")

plt.show()
````
