````python
# Funci√≥n que calcula el m√°ximo entre dos n√∫meros pasados
# como argumentos.
def numero_maximo(a, b):
    if a > b:
        maximo = a
    else:
        maximo = b
    return maximo
    
numero_maximo(10,1)    
````

### Funci√≥n Polin√≥micas

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una funci√≥n que tiene esta forma se le llama polinomio de grado $n$. A los elementos $a$ los llamaremoc coeficientes donde $a \in R$. 

**Ejemplo:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

que es un polin√≥mio de grado 7.


### Funciones potencia

Hay unas funciones que son un caso particular de las funciones polin√≥micas que son las funciones potencia, las cuales tienen la forma:


$$f(x)= x^{a}, a \in R$$ 


**Por ejemplo:**

$$f(x)= x^{2}$$

El dominio de $f(x)=x^{2}$ es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = [0, \infty)$

````python
import numpy as np               #Liberia para vectores
import matplotlib.pyplot as plt  # Libreria para
N = 40   #lo usaremos para el 


def f(x):
  return x**3-2*x**2-3*x+2

x = np.linspace(-10, 10, num = N)

y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y = 0, color = 'blue')
ax.axvline(x = 0, color = 'red')
plt.show()
````

# Exigencia de un argumento
Para requerir un argumento, agr√©galo entre par√©ntesis:

````python
def distancia_de_tierra(destino):
    if destino == 'Luna':
        return '238,855'
    else:
        return 'No se puede computar a ese destino'


distancia_de_tierra('Luna')
# distancia_de_tierra('Pluton') para el else

# Si colocamos otro argumento nos arroja el else
# si colocamos solo distancia_de_tierra no arrojara error
````
